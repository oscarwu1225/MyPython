```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
def vs (team1,team2):
    tm1=pd.read_html(team1)
    tm2=pd.read_html(team2)
    tm1data=np.array(tm1[3].iloc[0:7,3])
    tm2data=np.array(tm2[3].iloc[0:7,3])
    tm1score=np.zeros(7)
    tm2score=np.zeros(7)
    tm1sum=0
    tm2sum=0
    tm1win=0
    tm2win=0
    
    for i in range(7):
        
        for i in range(7):
            tm1score[i]=math.sqrt(tm1data[i])*(np.random.randn())+tm1data[i]
            tm2score[i]=math.sqrt(tm2data[i])*(np.random.randn())+tm2data[i]
        tm1sum=np.sum(tm1score)    
        tm2sum=np.sum(tm2score)    
        
        print(NAME(team1),int(tm1sum),':',int(tm2sum),NAME(team2))
    
        if tm1sum>tm2sum:
            tm1win=tm1win+1
        else:
            tm2win=tm2win+1
            
            
            
        if tm1win==4:
            print(NAME(team1),tm1win,':',tm2win,NAME(team2))
            print('\n')
            return (team1)
        
        elif tm2win==4:
            print(NAME(team1),tm1win,':',tm2win,NAME(team2))
            print('\n')
            return (team2)
```

```python
def NAME(web):
    web=web[42:45]
    if web[2]=='/':
        web=web[0:2]
    
    web=web.upper()
    return (web)
```

```python
BOS='http://www.espn.com/nba/team/stats/_/name/bos/seasontype/2'
BKN='http://www.espn.com/nba/team/stats/_/name/bkn/seasontype/2'
PHI='http://www.espn.com/nba/team/stats/_/name/phi/seasontype/2'
TOR='http://www.espn.com/nba/team/stats/_/name/tor/seasontype/2'
MIL='http://www.espn.com/nba/team/stats/_/name/mil/seasontype/2'
IND='http://www.espn.com/nba/team/stats/_/name/ind/seasontype/2'
ORL='http://www.espn.com/nba/team/stats/_/name/orl/seasontype/2'
DET='http://www.espn.com/nba/team/stats/_/name/det/seasontype/2'
GSW='http://www.espn.com/nba/team/stats/_/name/gs/seasontype/2'
DEN='http://www.espn.com/nba/team/stats/_/name/den/seasontype/2'
POR='http://www.espn.com/nba/team/stats/_/name/por/seasontype/2'
HOU='http://www.espn.com/nba/team/stats/_/name/hou/seasontype/2'
UTA='http://www.espn.com/nba/team/stats/_/name/utah/seasontype/2'
OKC='http://www.espn.com/nba/team/stats/_/name/okc/seasontype/2'
SAS='http://www.espn.com/nba/team/stats/_/name/sa/seasontype/2'
LAC='http://www.espn.com/nba/team/stats/_/name/lac/seasontype/2'
```

```python
print('\nRound 1: \n')
ER1_1=vs(MIL,DET)
ER1_2=vs(TOR,ORL)
ER1_3=vs(PHI,BKN)
ER1_4=vs(BOS,IND)
WR1_1=vs(GSW,LAC)
WR1_2=vs(DEN,SAS)
WR1_3=vs(POR,OKC)
WR1_4=vs(HOU,UTA)

print('\nRound 2: \n')
ER2_1=vs(ER1_1,ER1_4)
ER2_2=vs(ER1_2,ER1_3)
WR2_1=vs(WR1_1,WR1_4)
WR2_2=vs(WR1_2,WR1_3)

print('\nRound 3: \n')
ER3=vs(ER2_1,ER2_2)
WR3=vs(WR2_1,WR2_2)

print('\nFinal: \n')
champion=vs(ER3,WR3)
```
