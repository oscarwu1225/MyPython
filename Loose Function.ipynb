```python
%matplotlib inline
import math 
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def fy (x):
    y=3*x**2+6*x+18
    y=np.round(y,4)
    return(y)
def derfy (x):
    y=6*x+6
    y=np.round(y,4)
    return(y)
```

```python
x=np.linspace(-20,20,100)
plt.plot(x,fy(x))
plt.plot(x,derfy(x))
```

```python
x1=random.uniform(-20,20)
xtry=[x1]

while abs(derfy(x1))>0.01:
    x1=x1-0.01*derfy(x1)
    x1=np.round(x1,4)
    xtry.append(x1)
```

```python
xtry=np.array(xtry)
plt.plot(x,fy(x))
plt.plot(x,derfy(x))
plt.scatter(xtry,fy(xtry))
```
