```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
def fct (x,y):
    
    a=math.pow(math.e,-(x+1)**2-y**2)
    b=math.pow(math.e,-x**2-(y+1)**2)
    c=math.pow(math.e,-x**2-y**2)
    d=-x/5+x**3+y**5
    
    z=(1/3*a)-(3*(1-x)**2*b)-(20*c*d)
    
    return z
```

```python
def insert(x1,x2):
    a=random.uniform(0,10)
    b=random.uniform(0,10)
    return ((x1*a+x2*b)/(a+b))
```

```python
parents=[]
fit=[]
for i in range(8):
    x=random.uniform(-2,2)
    y=random.uniform(-2,2)
    z=fct(x,y)
    print(z)
    a=[x,y,z]
    parents.append(a)
    fit.append(((1/(z+18))**1.5))

parents=np.array(parents)
```

```python
prob=[fit[0]]
for i in range(7):
    a=prob[i]+fit[i+1]
    prob.append(a)
    
```

```python
prob
```

```python
matenumb=[]
for i in range(6):
    y=random.uniform(0,prob[7])
    for i2 in range(8):
        if y<prob[i2]:
            matenumb.append(i2)
            break 
```

```python
generationMin=[]
for i in range(100): 
    parents2=[]
    fit2=[]
    for i in range(8):
        p,o=random.sample(matenumb,2)
        print(p,o)
        if int(p)==int(o):
            x=parents[p][0]+random.uniform(-0.05,0.05)
            y=parents[p][1]+random.uniform(-0.05,0.05)
            if int(x)<-2:
                x=-2
            if int(x)>2:
                x=2
            if int(y)<-2:
                y=-2
            if int(y)>2:
                y=2
        else:
            a=parents[p][0]
            b=parents[o][0]
            c=parents[p][1]
            d=parents[o][1]
    
            x=insert(a,b)
            y=insert(c,d)
    
        print(x,y)
        z=fct(x,y)
        print(z)
    
        w=[x,y,z]
        parents2.append(w)
    
        fit2.append(((1/(z+18))**1.5))
    
    
    parents2=np.array(parents2)
    generationMin.append(min(parents[0:8,2]))
    row1,line1=np.where(parents==min(parents[0:8,2]))
    row2,line2=np.where(parents2==max(parents2[0:8,2]))

    parents2[row2]=parents[row1]


    prob2=[fit2[0]]
    for i in range(7):
            a=prob2[i]+fit2[i+1]
            prob2.append(a)
    
    
    matenumb=[]
    for i in range(6):
            y=random.uniform(0,prob2[7])
            for i2 in range(8):
                if y<prob2[i2]:
                    matenumb.append(i2)
                    break 
    
    
    
    parents=parents2
```
