```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
df=pd.read_html('http://www.stat-nba.com/player/195.html')
kbdt=df[1].iloc[0:20,6].str.strip("%").astype(float)
kbdt=np.array(kbdt)
```

```python
df=pd.read_html('http://www.stat-nba.com/player/1862.html')
lbjdt=df[1].iloc[0:16,6].str.strip("%").astype(float)
lbjdt=np.array(lbjdt)
```

```python
df=pd.read_html('http://www.stat-nba.com/player/1717.html')
mjdt=df[1].iloc[0:14,6].str.strip("%").astype(float)
mjdt=np.array(mjdt)
```

```python
plt.xlabel('Year')
plt.ylabel('Field Goal')
plt.plot(kbdt,label='Kobe')
plt.plot(lbjdt,label='LeBron')
plt.plot(mjdt,label='MJ')
plt.xticks(np.linspace(0,19,20))
plt.legend(loc='upper right')
```
